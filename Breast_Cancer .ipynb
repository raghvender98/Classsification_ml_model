```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter={"penalty":['l1', 'l2', 'elasticnet'],'C':[1,2,3,4,5,10,20,30,40,50,60],'max_iter':[100,200,300]}
```

```python
classifier_regressor=GridSearchCV(model,param_grid=parameter,scoring='accuracy',cv=5)
```

```python
classifier_regressor.fit(x_train, y_train)
```

```python
classifier_regressor.best_params_
```

```python
classifier_regressor.best_score_
```

```python
y_pred=classifier_regressor.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred, y_test)
```

```python
score
```

```python
print(classification_report(y_pred, y_test))
```
